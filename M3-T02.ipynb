# Anàlisi del dataset i les seves variables.

Sobre el dataset, aquest realitza un seguiment de la puntualitat dels vols nacionals als EUA operats per grans companyies aèries. La informació disponible analitza la quantitat de vols a temps, amb retard, cancel·lats i desviats de vols del 2008.

Es realitzarà una anàlisi del dataset proporcionat, el qual conté fins a 1.936.758 vols interns diferents als EUA per 2008 i les seves causes de retard, desviament i cancel·lació. Les dades provenen de la Oficina d'Estadístiques de Transport (BTS) del Departament de Transport dels EUA (DOT). Per treballar sobre ell, agafarem un sample de 200000 dades.

Les variables que hi podem trobar són les següents:

* **Year**: L'any, el 2008.

* **Month**: Mesos, de l'1 al 12.
* **DayofMonth**: Dia del mes, de l'1 al 31.
* **DayofWeek**: Dia de la setmana, de l'1 (dilluns) al 7 (diumenge).
* **DepTime**: hora de sortida real (local, hhmm).
* **CRSDepTime**: hora de sortida programada (local, hhmm).
* **ArrTime**: hora d'arribada real (local, hhmm).
* **CRSArrTime**: hora d'arribada programada (local, hhmm).
* **UniqueCarrier**: codi únic de la companyia aèria.
* **FlightNum**: número de vol.
* **TailNum**: número de cua de l'avió: matrícula de l'aeronau, identificador únic de l'aeronau.
* **ActualElapsedTime**: Temps real transcorregut en minuts.
* **CRSElapsedTime**: Temps transcorregut en minuts.
* **AirTime**: en minuts.
* **ArrDelay**: retard en l'arribada, en minuts: Un vol es compta com "a temps" si es va realitzar amb menys de 15 minuts de retard respecte a l'hora programada que es mostra en els Sistemes de Reserves Computaritzades (CRS en anglés) de les aerolínies.
* **DepDelay**: retard en la sortida, en minuts.
* **Origin**: codi d'origen de l'aeroport IATA.
* **Dest**: codi de destí d'aeroport IATA.
* **Distance**: Distància en milles.
* **TaxiIn**: Temps en minuts quan l'aeronau te autorització per entrar a pista.
* **TaxiOut**: Temps en minuts quan l'aeronau surt de pista.
* **Cancelled**: Si el vol va ser cancel·lat o no: 0 = no, 1 = sí.
* **CancellationCode**: motiu de la cancel·lació (A = transportista, B = clima, C = NAS, D = seguretat).
* **Diverted**: Desviats: 1 = sí, 0 = no.
* **CarrierDelay**: retard en minuts de l'aeronau per culpa de l'aerolínia. Per exemple: repostatge de l'aeronau, neteja de la mateixa, avaria, etc.
* **WeatherDelay**: retard en minuts a conseqüència de condicions climatològiques adverses que poden transcòrrer abans de la sortida de l'aeroport orígen, durant el vol, o a l'aeroport destí. Per exemple: tempestes elèctriques, grans nevades, etc.
* **NASDelay**: en minuts: el retard que està sota el control del Sistema Nacional de l'Espai Aèri (NAS) pot incloure: condicions climàtiques no extremes, operacions aeroportuàries, volum de tràfic pesat, control de tràfic aèri, etc.
* **SecurityDelay**: en minuts: el retard de seguretat es produeix per l'evacuació d'una terminal o vestíbul, tornar a embarcar una aeronau a causa d'una violació de la seguretat, equips de control inoperatius i/o llargues files de més de 29 minuts als controls de seguretat de l'aeroport.
* **LateAircraftDelay**: en minuts: Retard en l'arribada a un aeroport a causa de l'arribada tardana de la mateixa aeronau a un aeroport anterior. L'efecte dominó d'un retard anterior als aeroports posteriors es denomina propagació del retard.

* **TailNum**: Podriem fer un cribatge i comprovar si una mateixa aeronau acumula més retards que d'altres, depenent del model, antiguitat, possibles avaries més sovint, etc.

* **ActualElapsedTime**: Necessari per saber el temps real transcorregut.

* Les columnes **Cancelled** i **Diverted** per comprovar si hi ha hagut cancel·lació o desviació del vol.

* La columna **CancellationCode** per apreciar si el vol ha estat cancel·lat o no mitjançant el codi.

* **CarrierDelay**: Per veure el retard causat per l'aerolínia.

* **WeatherDelay**: Temps de retard causat per condicions climatològiques.

* **NASDelay**: Per comprovar el temps de retard causat per tancament d'espai aeri.

* **SecurityDelay**: Retard causat per problemes de seguretat.

* **LateAircraftDelay**: Ens és útil per saber si el retard ve causat perquè l'aeronau ja s'havia enlairat tard degut a un aterrament amb retard a l'aeroport orígen.

### - Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

```python
#Fem selecció de les dades que ens interessen:
df.loc[df['ArrDelay'] <= 15, 'Status'] = "On time"
df.loc[df['ArrDelay'] > 15, 'Status'] = "Delayed"
df.loc[df['Diverted'] == 1,  'Status'] = "Diverted"
df.loc[df['Cancelled'] == 1, 'Status'] = "Cancelled"

#Eliminem els no desitjats
df = df.drop(['Diverted'], axis=1)
df = df.drop(['Cancelled'], axis=1)

#Filtrem i mostrem els que porten retard:
df2 = df[df.Status != "On time"]
```

Segons el Sistema de Reserves Computaritzades (CRS) de les aerolínees, un vol arriba "a temps" si opera en menys de 15 minuts després de l'hora programada, llavors en base a això apliquem un condicional per saber l'status del vol. 

```python
import warnings
warnings.filterwarnings('ignore')

df2['Distance_km'] = df2['Distance'] * 1.60934 #Convertim milles a km.
df2['Vmean'] = df2['Distance_km'] / (df2['AirTime']/60) #Creem una nova columna amb la velocitat mitjana de l'aparell enlairat, en km/h.
```
