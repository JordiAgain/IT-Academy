# Exercici 1

En aquesta base de dades trobem 5 taules:

    ·Tb_genre
    ·Tb_movie
    ·Tb_role
    ·Tb_person
    ·Tb_movie_person
       
    


Dins aquestes taules a la taula *tb_genre* trobem dades dels gèneres de les pel·lícules, que en son 11. També podem trobar 5 columnes amb els atributs:

1. Genre_id, el qual és una clau primària per donar lloc al gènere.
2. Genre, el gènere per se.
3. Created_by_user. Usuari que l’ha creada.
4. Created_date. Data de creació.
5. Updated_date. Darrera data actualitzada.

Gràcies al diagrama E-R que hem fet, trobem la relació 1:N amb la taula tb_movie, el qual ens indica que una pel·lícula  pot pertànyer a un gènere en concret, així com alhora moltes pel·lícules poden pertànyer a un mateix gènere.

La taula tb_movie, presenta una relació 1:n amb la taula tb_movie_person, fent aquesta última de nexe entre les taules tb_person, tb_role i tb_Movie amb una relació n:n.  Alhora presenta una relació n:1 amb la taula tb_genre, ja que moltes pel·lícules poden pertànyer a un mateix gènere pero cada pel·lícula es d’un gènere.  Tornant a la taula tb_movie, presenta dades de diferents pel·lícules, amb 16 registres. Hi ha 8 columnes amb els següents atributs:

1. movie_id (primary key)
2. movie_title
3. movie_date
4. movie_format
5. movie_genre_id (foreign key)
6. created_by_user
7. created_date
8. updated_date
